```python
fil = '/home/ivo/S4/txt/pj_palavras_grams.txt'
fil = '/home/ivo/S4/txt/pj_palavras_todas.txt'
fil = '/home/ivo/S4/txt/st.txt'
#fil = '/home/ivo/Downloads/texto.csv'

target_docs = []
PYTHON_text = ''
with open(fil) as f:
    with tqdm(total=83959) as pbar:
        for i, linha in enumerate(f):
            PYTHON_text = PYTHON_text + linha.split(';')[1] + ' '
            target_docs.append(linha.split(';'))
            #print(i)
            pbar.update(1)
    #PYTHON_text = f.readline()
```

```python
import nltk

all_sentences = nltk.sent_tokenize(PYTHON_text)

all_words = [nltk.word_tokenize(sent) for sent in all_sentences]
```
